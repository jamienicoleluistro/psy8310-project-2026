### 3.3 Create an output location

```python
mkdir -p ~/PartA_output
```

### 3.4 Run fMRIprep

```python
# Load fMRIPrep
ml fmriprep/25.1.3

# Create a SUBJECTS_DIR (required for some BOLD↔T1 registration steps)
mkdir -p ~/PartA_output/freesurfer_subjects
export SUBJECTS_DIR=~/PartA_output/freesurfer_subjects

fmriprep ~/openneuro_data/ds005123 ~/PartA_output/fmriprep_out participant \
  --participant-label 11005 \
  --fs-license-file ~/.license \
  --fs-no-reconall \
  --skip-bids-validation \
  --output-spaces MNI152NLin6Asym \
  --nthreads 8 --omp-nthreads 2 --mem_mb 20000 \
  --work-dir ~/PartA_output/fmriprep_work
```
